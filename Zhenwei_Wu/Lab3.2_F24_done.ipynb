Another useful option is usecols. This lets you choose specific columns to load, which is helpful when you're working with large datasets and only need a few columns. This can save memory and processing time.

i.e: pd.read_csv('data.csv', index_col='ID', usecols=['ID', 'Name', 'Age'])

This code sets the "ID" column as the index and only loads the "ID", "Name", and "Age" columns from the file.

```python
# import data
data_sd = pd.read_csv('alldata_sed.csv', index_col = 'date')
data_sd
```

```python
subset_data = data_sd[['site', 'Salinity_mean']]
print(subset_data)
```

```python
# set up size of figure
plt.figure(figsize=(10, 6))

# Loop through each unique site and plot its data
for site in subset_data['site'].unique():
    site_data = subset_data[subset_data['site'] == site]
    plt.plot(site_data.index, site_data['Salinity_mean'], label=f'Site {site}')
    
# add title and label and legend and grid
plt.title('Salinity Over Time by Site')
plt.xlabel('Date')
plt.ylabel('Salinity Mean')
plt.legend(title="Site")
plt.grid(True)

# rotate x-label
plt.xticks(rotation=45)

# plt.tight_layout() is used in Matplotlib to automatically adjust the spacing 
# between subplots or plot elements (e.g., axes, labels, titles)  to prevent overlap and improve the overall layout.
plt.tight_layout()

plt.show()
```
